> tensor: ç±»ä¼¼äºå‘é‡ä½†æ¯”å‘é‡æ›´ä¸€èˆ¬çš„æ•°å­¦å¯¹è±¡ï¼Œç”±ç©ºé—´åæ ‡ç³»ä¸­çš„ä¸€ç»„å‡½æ•°è¡¨ç¤ºã€‚

é¦–å…ˆåŠ è½½ tensorflow å¹¶é‡ç½®è®¡ç®—å›¾

````python
import tensorflow as tf
from tensorflow.python.framework import ops
ops.reset_default_graph()
````

## å¯åŠ¨å›¾ä¼šè¯ï¼ˆgraph sessionï¼‰

ä½¿ç”¨ tf.Session() å–å¾—å›¾å¥æŸ„

````python
sess = tf.Session()
````

## åˆ›å»º Tensors

Tensorflow å¯ä½¿ç”¨å…§å»ºå‡½æ•°åˆ›å»ºç”¨äºå˜é‡çš„ tensorã€‚ä¾‹å¦‚ï¼Œä½¿ç”¨ tf.zeros() æ¥åˆ›å»ºé¢„å®šä¹‰å½¢çŠ¶çš„0å€¼ tensor

````python
my_tensor = tf.zeros([1, 20])
````

ä½¿ç”¨ run() æ–¹æ³•æ¥å¯¹ tensor è¿›è¡Œè¯„ä¼°

````python
sess.run(my_tensor)
````

tf ç®—æ³•éœ€è¦çŸ¥é“å“ªäº›å¯¹è±¡æ˜¯å¸¸é‡å“ªäº›æ˜¯å˜é‡ã€‚åœ¨æœ¬ç« åé¢ä¼šè§£é‡Šè¿™ä¸¤è€…çš„åŒºåˆ«ã€‚ç°åœ¨æˆ‘ä»¬ä½¿ç”¨ tf.Variable() åˆ›å»ºä¸€ä¸ªå˜é‡

````python
my_var = tf.Variable(tf.zeros([1, 20]))
````

æ³¨æ„ï¼Œç°åœ¨è¿˜ä¸èƒ½è¿è¡Œ sess.run(my_var)ï¼Œå› ä¸º tf ä½¿ç”¨è®¡ç®—å›¾è¿›è¡Œè®¡ç®—ï¼Œä¸ºäº†è®¡ç®—å˜é‡æˆ‘ä»¬å¾—å…ˆå¯¹å˜é‡è¿›è¡Œåˆå§‹åŒ–æ“ä½œã€‚åé¢ä¼šçœ‹åˆ°æ›´å¤šçš„è¿™ç§æ“ä½œã€‚åœ¨è¿™é‡Œï¼Œæˆ‘ä»¬ä½¿ç”¨ my_var.initializer åˆå§‹åŒ–ä¸€ä¸ªå˜é‡

````python
sess.run(my_var.initializer)
sess.run(my_var)
````

ä¸ºåˆ›å»ºç‰¹å®šå½¢çŠ¶çš„å˜é‡ï¼Œè®©æˆ‘ä»¬å…ˆå£°æ˜è¡Œåˆ—çš„å¤§å°

````python
row_dim = 2
col_dim = 3
````

ç„¶åä½¿ç”¨0å€¼å’Œ1å€¼å‡½æ•°å£°æ˜è¡Œå’Œåˆ—å˜é‡

````python
zero_var = tf.Variable(tf.zeros([row_dim, col_dim]))
ones_var = tf.Variable(tf.ones([row_dim, col_dim]))
````

åŒæ ·çš„ï¼Œä½¿ç”¨åˆå§‹åŒ–å‡½æ•°åˆå§‹åŒ–å˜é‡ï¼Œç„¶å run æ¥èµ‹å€¼å®ƒä»¬çš„å†…å®¹

````python
sess.run(zero_var.initializer)
sess.run(ones_var.initializer)
print(sess.run(zero_var))
print(sess.run(ones_var))
````

## æ‹·è´å…¶ä»– tensor çš„å½¢çŠ¶æ¥åˆ›å»ºæ–° tensor

å¯ä»¥ä½¿ç”¨å…§å»ºå‡½æ•° ones_like() æˆ– zeros_like() æ¥æ‹·è´å…¶ä»– tensor çš„å½¢çŠ¶

````python
zero_similar = tf.Variable(tf.zeros_like(zero_var))
ones_similar = tf.Variable(tf.ones_like(ones_var))

sess.run(zero_similar.initializer)
sess.run(ones_similar.initializer)
print(sess.run(zero_similar))
print(sess.run(ones_similar))
````

## ä½¿ç”¨å¸¸é‡å¡«å…… tensor

ğŸ‘‡ä½¿ç”¨å¸¸é‡å¡«å…… tensor

````python
fill_var = tf.Variable(tf.fill([row_dim, col_dim], -1))
sess.run(fill_var.initializer)
sess.run(fill_var)
````

æˆ‘ä»¬ä¹Ÿå¯ä»¥ä»å¸¸é‡æ•°ç»„æˆ–åˆ—è¡¨åˆ›å»º tensor å˜é‡

````python
# ä»å¸¸é‡åˆ›å»º tensor å˜é‡
const_var = tf.Variable(tf.constant([8, 6, 7, 5, 3, 0, 9]))
# ä¹Ÿå¯ç”¨äºå¡«å……æ•°ç»„
const_fill_var = tf.Variable(tf.constant(-1, shape=[row_dim, col_dim]))

sess.run(const_var.initializer)
sess.run(const_fill_var.initializer)

print(sess.run(const_var))
print(sess.run(const_fill_var))
````

## ä»åºåˆ—å’Œ Ranges ä¸­åˆ›å»º tensor

æˆ‘ä»¬ä¹Ÿå¯ä»¥ä»åºåˆ—ç”Ÿæˆå‡½æ•°ç”Ÿæˆ tensorã€‚tf çš„ linspace() å‡½æ•°å’Œ range() å‡½æ•°å’Œ python/numpy çš„æ“ä½œéå¸¸ç›¸ä¼¼ã€‚

````python
# tf çš„ linspace
linear_var = tf.Variable(tf.linspace(start=0.0, stop=1.0, num=3))

# tf çš„ range
sequence_var = tf.Variable(tf.range(start=6, limit=15, delta=3))

sess.run(linear_var.initializer)
sess.run(sequence_var.initializer)
print(sess.run(linear_var))
print(sess.run(sequence_var))
````

## éšæœºæ•°å­— tensor

æˆ‘ä»¬ä¹Ÿå¯ä»¥å¦‚ä¸‹ä»éšæœºæ•°å­—ä¸­åˆå§‹åŒ– tensor

````python
rnorm_var = tf.random_normal([row_dim, col_dim], mean=0.0, stddev=1.0)
runif_var = tf.random_uniform([row_dim, col_dim], minval=0, maxval=4)
print(sess.run(rnorm_var))
print(sess.run(runif_var))
````

## åœ¨ TensorBoard ä¸­å¯è§†åŒ–å˜é‡åˆ›å»ºè¿‡ç¨‹

ä¸ºäº†èƒ½å¯è§†åŒ–å˜é‡çš„åˆ›å»ºè¿‡ç¨‹ï¼ˆç¬¬11ç« æœ‰æ›´å¤šç»†èŠ‚ï¼‰ï¼Œæˆ‘ä»¬å°†é‡ç½®è®¡ç®—å›¾å¹¶åˆ›å»ºä¸€ä¸ªå…¨å±€çš„åˆå§‹åŒ–æ“ä½œ

````python
# é‡ç½®å›¾
ops.reset_default_graph()

# é‡å¯åŠ¨ä¼šè¯
sess = tf.Session()

# åˆ›å»ºå˜é‡
my_var = tf.Variable(tf.zeros([1, 20]))

# æ·»åŠ æ¦‚è¦åˆ° tensorboard
merged = tf.summary.merge_all()

# åˆå§‹åŒ–å›¾ writer
writer = tf.summary.FileWriter('./variable_logs', graph=sess.graph)

# åˆå§‹åŒ–æ“ä½œ
initialize_op = tf.global_variables_initializer()

# è¿è¡Œâ€œåˆå§‹åŒ–æ“ä½œâ€
sess.run(initialize_op)
````
